#### Задача 10.1.10
$$
\begin{cases}
-\frac{d}{dx} \left( K(x) \frac{du}{dx} \right) = f(x) \\
u(a) = UA,\hspace{0.5cm} u(b) = UB
\end{cases}
$$
где
$$
k(x) = x,\hspace{0.3cm} a=0.1,\ b=0.6, \hspace{0.3cm}UA = 1,\ Ub = 5  
$$

```python
from sympy import Symbol, Function, Derivative
from sympy import dsolve, Eq
from sympy import ln

x = Symbol('x')
u = Function('u')(x)

f = ln(x)
k = x

a, b = 0.1, 0.6
Ua, Ub = 1.0, 5

sets = [
    (k, Ua, Ub),
    (10 * k, Ua, Ub),
    (0.1 * k, Ua, Ub),
    (1 / k, Ua, Ub),
    (k, -Ua, Ub),
    (k, Ua, -Ub),
    (k, -Ua, -Ub)
]
```

```python
from sympy import lambdify


x_arr = np.linspace(a, b, 150)
solutions = {}


for i, (K, Ua, Ub) in enumerate(sets, start=1):
    diff_eq = Eq(- Derivative(K * u.diff(x), x), f)
    
    ics = {
        u.subs(x, a): Ua,
        u.subs(x, b): Ub
    }

    u_sol = dsolve(diff_eq, u, ics=ics).rhs
    u_func = lambdify(x, u_sol, 'numpy')

    solutions[i] = u_func(x_arr)
```

```python
plt.plot(x_arr, solutions[1], label='1st paramaters set')
plt.plot(x_arr, solutions[2], label='2nd paramaters set')
plt.plot(x_arr, solutions[3], label='3rd parameters set')


plt.legend()
plt.savefig('plots1/sets1-3.png', format='png')
plt.grid()
plt.show()
```

#### Решение СЛАУ с трёхдиагональной матрицей методом прогонки.

```python
def tridiagonal_solver(l, m, u, r):
    """
    Solves a system of linear equations with a tridiagonal matrix by the run-through method.

    :param l np.ndarray: the lower diagonal of matrix.
    :param: np.ndarray m: the main diagonal of matrix.
    :param: np.ndarray u: the upper diagonal of matrix.
    :param: np.ndarray r: the right size of system.

    :return np.ndarray: vector with solution of system
    """
    n = m.size       # l.size = u.size = n - 1
    a = np.zeros(n)
    b = np.zeros(n)

    a[0] = -u[0] / m[0]
    b[0] = r[0] / m[0]

    for i in range(1, n-1):
       a[i]=-u[i] / (l[i-1] * a[i-1] + m[i])
       b[i]=(r[i] - l[i-1] * b[i-1]) / (l[i-1] * a[i-1] + m[i])

    X = np.zeros(n)
    b[n - 1]= (r[n - 1] - l[n - 2] * b[n - 2])/(m[n - 1] + l[n - 2] * a[n - 2])
    X[n - 1] = b[n - 1]

    for i in range(n - 2, -1, -1):
        X[i] = a[i]*X[i+1] + b[i]
    return X
```

#### Задача 10.3.5

$$
\begin{cases}
u''(x) - u'(x) + 2 x^2 u(x) = x + 1 \\
u(1.3) = 1\\
u(2.4) + u'(2.4) = 3.2
\end{cases}
$$

Обозначим $U_a = 1$, $U_b = 3.2$, тогда система линейных уравнений будет иметь вид:
$$
\begin{cases}
u_0 = 1 \\
(2 - h) u_{i+1} + (-4 + 4x_i^2 h^2) u_i + (2 + h)u_{i-1} = 2h^2 (x_i + 1) \hspace{0.5cm} 1 \le i \le n \\
(2h^2 + 8h + 4) u_n + (-4h -4 -4 x_{n-1}^2 h^2) u_{n-1} = 2 * h * U_b (h + 2) - 2 h^2 (x_{n-1} + 1)
\end{cases}
$$

```python
def get_solution3(n):
    a = 1.3
    b = 2.4

    Ua = 1
    Ub = 3.2

    h = (b - a) / n
    x_arr = np.linspace(a, b, n + 1)


    lower = np.zeros(n - 1)
    main = np.zeros(n)
    upper = np.zeros(n - 1)
    b = np.zeros(n)

    
    for i in range(n):
        x = x_arr[i+1]
    
        if i == 0:
            main[i] = -4 + 4 * (x ** 2) * (h ** 2)
            upper[i] = 2 - h
            b[i] = 2 * (h ** 2) * (x + 1) - (2 + h) * Ua
    
        elif 0 < i and i < n-1:
            lower[i - 1] = 2 + h
            main[i] = -4 + 4 * (x ** 2) * (h ** 2)
            upper[i] = 2 - h
            b[i] = 2 * (h ** 2) * (x + 1)
        else:
            x_prev = x_arr[-2]

            lower[i - 1] = -4 * h - 4 - 4 * (x_prev ** 2) * (h ** 2)
            main[i] = 2 * (h ** 2) + 8 * h + 4
            b[i] = 2 * h * Ub * (h + 2) - 2 * (h ** 2) * (x_prev + 1)


    y_arr = np.zeros(n + 1)
    y_arr[0] = Ua
    y_arr[1:] = tridiagonal_solver(lower, main, upper, b)

    return x_arr, y_arr
```

```python
x_arr1, y_arr1 = get_solution3(100)
x_arr2, y_arr2 = get_solution3(200)

plt.plot(x_arr1, y_arr1, color='purple', label='$n=150$')
plt.plot(x_arr2, y_arr2, color='red', label='$n=300$')

plt.title("Numerical solution")
plt.grid()
plt.legend()

plt.savefig("plots2/solutions.png", format='png')
plt.show()
```

```python
def runge_error(y1, y2, p=2):
    return np.abs(y_arr1 - y_arr2[::2]).max() / (2 ** p - 1)

eps = 0.1
rate = runge_error(y_arr1, y_arr2)

print(f"Error: {rate}")
print(f"{'Error is less eps!' if rate < eps else 'Error is more than required!'}")
```

#### Задача 10.4.10
Исходное уравнение:
$$
-\frac{d}{dx} \left( k(x) \frac{du}{dx} \right) = f
$$
Или по-другому
$$
-k'(x) u'(x) - k(x) u''(x) = f(x)
$$
Разностная схема для такого уравнения будет иметь следующий вид:
$$
-k'_i \frac{u_{i+1} - u_{i-1}}{2h} - k_i \frac{u_{i+1} - 2u_i + u_{i-1}}{h^2} = f_i
$$

Система линейных уравнений:
$$
\begin{cases}
u_0 = Ua, \hspace{0.4cm} u_n = Ub \\ 
(-h k'_i - 2 k_i) u_{i+1} + 4 k_i u_i + (h k'_i - 2k_i) u_{i-1} = 2h^2 f_i\hspace{0.5cm} 1 \le i \le n-1 \\
\end{cases}
$$

##### 2.1 Данные из задачи 10.1.10
$$
\begin{cases}
k(x) = x,\ f(x) = \ln(x) \\
a = 0.1,\ Ua = 1 \\
b = 0.6,\ Ub = 5 \\
\end{cases}
$$

```python
def get_solution4(n, k_func, k_der_func, f_func, a, b, Ua, Ub):
    h = (b - a) / n
    
    x_arr = np.linspace(a, b, n + 1)
    
    lower = np.zeros(n - 2)
    main = np.zeros(n - 1)
    upper = np.zeros(n - 2)
    b_arr = np.zeros(n - 1)

    x_arr = np.linspace(a, b, n + 1)
    
    for i in range(n-1):
        xi = x_arr[i+1]
        ki = k_func(xi)
        dki = k_der_func(xi)
        fi = f_func(xi)
    
        if i == 0:
            main[i] = 4 * ki
            upper[i] = -h * dki - 2 * ki
            b_arr[i] = 2 * (h ** 2) * fi - Ua * (h * dki - 2 * ki)
    
        elif 0 < i and i < n-2:
            lower[i-1] = (h * dki - 2 * ki)
            main[i] = 4 * ki
            upper[i] = -h * dki - 2 * ki
            b_arr[i] = 2 * (h ** 2) * fi
        else:
            lower[i - 1] = (h * dki - 2 * ki)
            main[i] = 4 * ki
            b_arr[i] = 2 * (h ** 2) * fi - Ub * (-h * dki - 2 * ki)


    y_arr = np.zeros(n + 1)
    y_arr[0] = Ua
    y_arr[-1] = Ub
    y_arr[1:-1] = tridiagonal_solver(lower, main, upper, b_arr)
    return x_arr, y_arr
```

```python
def f(x):
    return np.log(x)

a, b = 0.1, 0.6
Ua, Ub = 1, 5
n = 150
step = (Ub - Ua) / 150
```

```python
def k_func_1(x):
    return x


def k_func_der_1(x):
    return 1.0


x_arr, y_arr = get_solution4(n, k_func_1, k_func_der_1, f, a, b, Ua, Ub)

plt.plot(x_arr, y_arr, c='springgreen')
plt.grid()
plt.savefig('plots4/task1.png', format='png')
plt.show()
```

#### Задача 10.5.10

##### Условие задачи:
$$
\begin{cases}
-(k(x) u'(x))' + q(x)u(x) = f(x),\ \ x \in (a, b) \\
-k(a)u'(a) + 0.5 u(a) = 0 \\
k(b)u'(b) + 0.5 u(b) = 0
\end{cases}
$$
где
$$
k(x) = \begin{cases}
1.8,\ \ a < x < c \\
0.6,\ \ c < x < b
\end{cases}
$$
$$
q(x) = \begin{cases}
6.5,\ \ a < x < c\\
7.8,\ \ c < x < b
\end{cases}
$$
$$
f(x) = 8x (2.5 - x)
$$
$$
a=0,\ b=2,\ c=1.125
$$

##### Система линейных уравнений:
$$
\begin{cases}
(-2 k_1 k_0 + h^2 k_0 q_1)u_1 + (2 k_0 k_1 + k_1 h)u_0 = k_0 f_1 h^2 \\
-k_i u_{i+1} + (2k_i + h^2 q_i) u_i - k_i u_{i-1} = f_i h^2,\ \ \ 1 \le i \le n-1 \\
(2 k_n k_{n-1} + h k_{n-1})u_n + (-2k_n k_{n-1} + k_n h^2 q_{n-1})u_{n-1} = k_n f_{n-1} h^2
\end{cases}
$$

Замечание: система выше была выведена с учётом того, что $k'(x) = 0$

```python
a, b, c = 0, 2, 1.125

def k(x):
    return 1.8 if x < c else 0.6


def q(x):
    return 6.5 if x < c else 7.8


def f(x):
    return 8 * x * (2.5 - x)


def get_solution5(n):
    h = (b - a) / n
    x_arr = np.linspace(a, b, n+1)

    lower = np.zeros(n)
    main = np.zeros(n+1)
    upper = np.zeros(n)
    b_vec = np.zeros(n+1)

    for i in range(n+1):
        if i == 0:
            x = x_arr[i]
            x_next = x_arr[i+1]

            k0 = k(x)
            k1 = k(x_next)

            q1 = f(x_next)
            f1 = f(x_next)
            
            main[i] = (2 * k0 * k1 + k1 * h)
            upper[i] = (-2 * k1 * k0 + (h ** 2) * k0 * q1)
            b_vec[i] = (k0 * f1 * (h ** 2))
            
        elif i < n:
            x = x_arr[i]
            ki = k(x)
            qi = q(x)
            fi = f(x)

            lower[i-1] = (-ki)
            main[i] = (2 * ki + (h ** 2) * qi)
            upper[i] = (-ki)
            b_vec[i] = fi * (h ** 2)

        elif i == n:
            x = x_arr[i]
            x_prev = x_arr[i-1]

            kn = k(x)
            kp = k(x_prev)
            qp = q(x_prev)
            fp = f(x_prev)
            
            lower[i-1] = (-2 * kn * kp + kn * qp * (h ** 2))
            main[i] = (2 * kn * kp + h * kp)
            b_vec[i] = kn * fp * (h ** 2)

    y_arr = tridiagonal_solver(lower, main, upper, b_vec)
    return x_arr, y_arr
```

```python
x_arr, y_arr = get_solution5(300)

plt.plot(x_arr, y_arr, color='deepskyblue', label='$u(x)$')

plt.grid()
plt.legend()

plt.savefig('plots5/problem5.png', format='png')
plt.show()
```

```python
x_arr1, y_arr1 = get_solution5(300)
x_arr2, y_arr2 = get_solution5(600)

rate = runge_error(y_arr1, y_arr2)
print(f"Error rate: {rate}")
```

#### Задача 10.6.10

##### Условие задачи
$$
\begin{cases}
\frac{\partial{u}}{\partial{t}} =
\frac{\partial}{\partial{x}} \left(
k(x) \frac{\partial{u}}{\partial{x}}
\right) + f(x) (1 - e^{-t}),\ \ \ \ 0 < x < l,\ \ \ \ 0 < t < T \\
u(0, t) = UA, \ \ \ u(l, t) = UB, \ \ \ \ 0 \le t \le T \\
u(x, 0) = \phi(x), \ \ \ \ 0 \le x \le l
\end{cases}
$$
где
$$ k(x) = x,\hspace{0.5cm} f(x) = \ln(x),\hspace{0.5cm} \phi(x) = \frac{UB - UA}{b - a}(x - a) + UA $$
$$ a = 0.1,\ UA = 1,\ b = 0.6,\ UB = 5 $$

При этом здесь и далее будем использовать дополнительное обозначение:
Введём обозначение
$$
g(x, t) = f(x)(1 - e^{-t})
$$


```python
import matplotlib.animation as animation
from IPython.display import HTML, Video, clear_output
%matplotlib inline
```

```python
a, b = 0.1, 0.6
ua, ub = 1.0, 5.0
tau = 0.05
h = 0.1


def k(x):
    return x

def k_der(x):
    return 1.0

def g(x, t):
    return np.log(x) * (1 - np.exp(-t))

def phi(x):
    return (x - a) * (ub - ua) / (b - a) + ua
```

##### Явная схема

$$
u^{j+1}_{i} = 
\tau \left( \frac{k'_i}{2h} + \frac{k_i}{h^2} \right) u^{j}_{i+1} + 
\left( 1 - \frac{2 \tau k_i}{h^2} \right) u^{j}_{i} + 
\tau \left( -\frac{k'_i}{2h} + \frac{k_i}{h^2} \right) u^{j}_{i-1} + 
\tau g^j_{i}
$$

```python
class ExplicitNPDESolution:
    """
    Represents numerical solution of PDE using explicit scheme
    """
    def __init__(self, k_func, k_der_func, g_func, a, b, ua, ub, phi_func, tau, h):
        self.n = int((b - a) / h)
        self.time = 0
        self.tau = tau
        self.h = h
        self.bounds = (a, b)
        self.bc = (ua, ub)

        self.x_list = np.linspace(a, b, self.n + 1)

        self.ki_list = np.array([k_func(x) for x in self.x_list])
        self.dki_list = np.array([k_der_func(x) for x in self.x_list])
        self.y_list = np.array([phi_func(x) for x in self.x_list])

        self.g_function = g_func

    def get_solution(self, return_time=False):
        if return_time:
            return self.x_list, self.y_list, self.time
        return self.x_list, self.y_list
    
    def update(self):
        time = self.time
        ua, ub = self.bc
        tau, h = self.tau, self.h

        y_now = self.y_list
        y_next = np.zeros_like(self.y_list)
        y_next[0], y_next[-1] = ua, ub

        dki = self.dki_list
        ki = self.ki_list
        gi = np.array([self.g_function(x, time) for x in self.x_list])
        
        ai = tau * (dki[1:-1] / (2 * h) + ki[1:-1] / (h ** 2))
        bi = (1 - (2 * tau * ki[1:-1]) / (h ** 2))
        ci = tau * (- dki[1:-1] / (2 * h) + ki[1:-1] / (h ** 2))
        di = tau * gi[1:-1]

        y_next[1:-1] = ai * y_now[2:] + bi * y_now[1:-1] + ci * y_now[:-2] + di
        
        self.time += self.tau
        self.y_list = y_next
```

```python
epde = ExplicitNPDESolution(
    k_func=k, k_der_func=k_der, g_func=g,
    a=a, b=b, ua=ua, ub=ub, phi_func=phi,
    tau=tau, h=h
)

M = 100

for i in range(M):
    epde.update()

x, y, t = epde.get_solution(return_time=True)

plt.plot(x, y, color='fuchsia', label=f't = {t:.2f}')
plt.grid()
plt.legend()

plt.savefig('plots6/explicit-failure.png', format='png')
plt.show()
```

##### Неяная схема

$$
\tau \left( \frac{k'_i}{2h} + \frac{k_i}{h^2} \right) u_{i+1}^{j+1} +
\left( - 1 - \frac{2 \tau k_i}{h^2} \right) u_{i}^{j+1} + 
\tau \left( \frac{k_i}{h^2} - \frac{k'_i}{2h} \right) u_{i-1}^{j+1} =
-u_i^j - \tau g_i^{j+1}
$$

```python
class ImplicitNPDESolution:
    """
    Represents numerical solution of PDE using implicit scheme
    """
    def __init__(self, k_func, k_der_func, g_func, a, b, ua, ub, phi_func, tau, h):
        self.n = int((b - a) / h)
        self.time = 0
        self.tau = tau
        self.h = h
        self.bounds = (a, b)
        self.bc = (ua, ub)

        self.x_list = np.linspace(a, b, self.n + 1)

        self.ki_list = np.array([k_func(x) for x in self.x_list])
        self.dki_list = np.array([k_der_func(x) for x in self.x_list])
        self.y_list = np.array([phi_func(x) for x in self.x_list])

        self.g_function = g_func

    def get_solution(self, return_time=False):
        if return_time:
            return self.x_list, self.y_list, self.time
        return self.x_list, self.y_list

    def update(self):
        time = self.time
        ua, ub = self.bc
        tau, h = self.tau, self.h

        y_next = np.zeros_like(self.y_list)
        y_next[0], y_next[-1] = ua, ub

        upper = np.empty(0)
        lower = np.empty(0)
        main = np.empty(0)
        vector = np.empty(0)

        for i in range(1, self.n):
            xi = self.x_list[i]
            ki = self.ki_list[i]
            dki = self.dki_list[i]
            ui = self.y_list[i]
            gi = self.g_function(xi, time + tau)

            ai = tau * (dki / (2 * h) + ki / (h ** 2))
            bi = -1 - (2 * tau * ki) / (h ** 2)
            ci = tau * (ki / (h ** 2) - dki / (2 * h))
            di = -ui - tau * gi

            if i == 1:
                upper = np.append(upper, ai)
                main = np.append(main, bi)
                vector = np.append(vector, di - ci * y_next[0])
            elif i < self.n - 1:
                upper = np.append(upper, ai)
                main = np.append(main, bi)
                lower = np.append(lower, ci)
                vector = np.append(vector, di)
            elif i == self.n - 1:
                main = np.append(main, bi)
                lower = np.append(lower, ci)
                vector = np.append(vector, di - ai * y_next[-1])

        y_next[1:-1] = tridiagonal_solver(lower, main, upper, vector)
        
        self.y_list = y_next
        self.time += self.tau          
```

```python
ipde = ImplicitNPDESolution(
    k_func=k, k_der_func=k_der, g_func=g,
    a=a, b=b, ua=ua, ub=ub, phi_func=phi,
    tau=tau/2, h=h
)

fig, ax = plt.subplots()

def init_plot():
    ax.clear()
    x_arr, y_arr, time = ipde.get_solution(return_time=True)
    
    plt.plot(x_arr, y_arr, color='purple', label=f't = {time:.3f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()


def update_plot(frame):
    ax.clear()

    x_arr, y_arr, time = ipde.get_solution(return_time=True)
    ipde.update()
    
    ax.plot(x_arr, y_arr, color='purple', label=f't = {time:.3f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)
    ax.legend()

    if np.isclose(time, [0.5 * tau, 10 * tau, 20 * tau]).any():
        plt.savefig(f'plots6/implicit-time{time:.3f}.png', format='png')


ani = animation.FuncAnimation(
    fig, update_plot, frames=400, init_func=init_plot
)

FFwriter = animation.FFMpegWriter(fps=10)
ani.save('plots6/implicit-animation.mp4', writer=FFwriter)

plt.show()
clear_output()

Video('plots6/implicit-animation.mp4')
```
